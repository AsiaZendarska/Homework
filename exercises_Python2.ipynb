5. Zakładamy, że prowadzisz analizę wydatków w swoim domowym budżecie. Masz informacje na temat procentowego udziału wydatków w kilku kategoriach. Twoim zadaniem jest stworzenie wykresu kołowego, który przedstawi te udziały procentowe. Poniżej znajdziesz dane dotyczące procentowego udziału w trzech kategoriach: Jedzenie: 35%, Mieszkanie: 40%, Rozrywka: 25%
Treść Zadania: Stwórz wykres kołowy przedstawiający procentowy udział wydatków w trzech kategoriach, Dodaj etykiety dla każdej kategorii na wykresie, Skonfiguruj kolor kawałków wykresu, aby były łatwo rozróżnialne, Dodaj procentowe etykiety do kawałków wykresu.

```python
kategorie = ['Jedzenie', 'Mieszkanie', 'Rozrywka']
udziały = [35, 40, 25]

fig, ax = plt.subplots()
ax.pie(udziały, labels=kategorie, autopct='%1.1f%%')
```

```python
kolory = ['#ff9999','#66b3ff','#99ff99']

plt.figure(figsize=(4, 4))
plt.pie(udziały, labels=kategorie, colors=kolory, autopct='%1.1f%%', startangle=140)
plt.title('Procentowy Udział Wydatków')
plt.axis('equal')  # Zapewnia, że koło jest równo okrągłe
plt.show()
```

6. Załaduj zbiór danych "flights" z biblioteki Seaborn. Przygotuj histogram dla kolumny "passengers". Dostosuj liczbę przedziałów (bins) i kolor histogramu. Dodaj tytuł wykresu oraz opisy osi. Wyświetl ostateczny histogram.

```python
flights = sns.load_dataset("flights")

plt.figure(figsize=(8, 6))
sns.histplot(flights['passengers'], bins=20, color='skyblue')
plt.title('Liczba Pasażerów w Przedziałach Czasowych')
plt.xlabel('Liczba Pasażerów')
plt.ylabel('Częstotliwość')
plt.show()
```

7. Przygotuj violin plot, aby zobaczyć rozkład długości płetw pingwinów w zależności od ich gatunku. Załaduj zbiór danych "penguins" z biblioteki Seaborn. Przygotuj violin plot dla kolumny "species" (gatunek) na osi X i "flipper_length_mm" (długość płetwy) na osi Y. Dostosuj paletę kolorów, aby podkreślić różnice między gatunkami. Dodaj tytuł wykresu i opisy osi. Wyświetl ostateczny violin plot.

```python
penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8, 6))
sns.violinplot(x="species", y="flipper_length_mm", data=penguins, hue="species", palette="muted", legend=False)
plt.title('Rozkład długości płetw pingwinów w zależości od och gatunku')
plt.xlabel('Gatunek')
plt.ylabel('Długość płetw')
plt.show()
```
